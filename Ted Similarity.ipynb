```python
import pandas as pd
import numpy as np
```

```python
people = pd.read_csv('ted_merged.csv')
```

```python
people = people.dropna()
```

```python
# Data contains: link to wikipedia article, name of person, text of article.
people.head()
```

```python
Ken = people[(people.main_speaker == 'Ken Robinson') & (people.event == 'TED2006')]

```

```python
Ken
```

```python
Ken['transcript'].values
```

```python
#Get the word counts for article
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
features = vectorizer.fit_transform(Ken['transcript'])
```

```python
features
```

```python
vocab = vectorizer.get_feature_names()
```

```python
vocab
```

```python
dist = np.sum(features.toarray(), axis=0)
dist
```

```python
#Sort the word counts for the article
#Turning dictonary of word counts into a dataframe

Ken_word_count_table = pd.DataFrame({'token':vocab, 'count':dist}).sort_values('count')
```

```python
Ken_word_count_table.sort_values('count', ascending=False).head(15)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer()
```

```python

dtm = vect.fit_transform(people['transcript'])
```

```python
dtm.shape
```

```python
features = vect.get_feature_names()
```

```python
features
```

```python
#Examine the TF-IDF for the article
#Remember that the  article is at index 35817
Ken
```

```python
#So we'll examine the document term matrix at index 35817 and pull out the TF-IDF by each word.
#First we'll make a dictionary
word_scores = {}
for word in vocab:
    word_scores[word] = dtm[0, features.index(word)]
```

```python
Ken_word_scores = pd.DataFrame(list(word_scores.items()),
                      columns=['word', 'tfidf'])
```

```python
Ken_word_scores.sort_values('tfidf', ascending=False).head(15)
```

```python
#Is Obama closer to Clinton than to Beckham?
#Note that we're computing the cosine similarity (which ranges from 0 to 1, where 1 is most similar) not the cosine distance which we can use as well.
#cosine_distance = (1-cosine_similarity)
#We should find that the article about president Obama is closer to the one about former president Clinton than that of footballer David Beckham.


from sklearn.metrics.pairwise import cosine_similarity
clinton= people[people['main_speaker'] == 'Al Gore']
beckham = people[people['main_speaker'] == 'Hans Rosling']
```

```python
cosine_similarity(dtm[Ken.index], dtm[clinton.index])
```

```python
cosine_similarity(dtm[Ken.index], dtm[beckham.index])
```

```python
#cosine distance
from sklearn.metrics.pairwise import pairwise_distances
pairwise_distances(dtm[Ken.index], dtm[clinton.index], metric='cosine')
```

```python
pairwise_distances(dtm[Ken.index], dtm[beckham.index], metric='cosine')
```

```python
#Build a nearest neighbor model for document retrievalÂ¶
dtm.shape
```

```python
from sklearn.neighbors import NearestNeighbors
neighbors = NearestNeighbors(n_neighbors=10)
```

```python
neighbors.fit(dtm)
```

```python
#Who is closest to Obama?
distances, indices = neighbors.kneighbors(dtm[Ken.index], return_distance=True)
```
